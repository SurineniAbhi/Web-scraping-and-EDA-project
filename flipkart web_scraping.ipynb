```python
# Impoting packages
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import regex as re
```

```python
# Web Scraping form Flipkart 
modelname=[]
discountprice=[]
originalprice=[]
powerconsumption=[]
reviews=[]
rating=[]
capacity=[]
#iterate through 24 pages
for i in range(1,25):
    url="https://www.flipkart.com/search?q=air+conditioners&sid=j9e%2Cabm%2Cc54&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_13_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_13_na_na_na&as-pos=1&as-type=RECENT&suggestionId=air+conditioners%7CAir+Conditioners&requestId=3340a220-fa71-4ca1-9a3a-812834786172&as-backfill=on&page="+str(i)
    page=requests.get(url)
    pagecontent=page.text
    soup = BeautifulSoup(pagecontent)
    
    cap=soup.find_all("div",attrs={'class':'_4rR01T'})
    for x in cap:
        l=x.text
        m=re.findall("[+-]?[0-9]+\.[0-9]+ | [0-9] Ton",l)
        capacity.append(m)
    
    for x in soup.find_all("div",attrs={'class':'_3pLy-c row'}):
        mname = x.find('div', attrs = {'class':'_4rR01T'})
        dprice = x.find('div', attrs = {'class':'_30jeq3 _1_WHN1'})
        oprice=x.find('div', attrs = {'class':'_3I9_wc _27UcVY'})
        powerconsump = x.find('li', attrs = {'class':'rgWa7D'})
        emi1 = x.find('div', attrs = {'class':'_3xFhiH'})
        rewi= x.find('span', attrs = {'class':'_2_R_DZ'})
        score = x.find('div', attrs = {'class':'_3LWZlK'})
        
        if mname is None:
            modelname.append(np.NaN)
        else:
            modelname.append(mname.text)
        
        if dprice is None:
            discountprice.append(np.NaN)
        else:
            discountprice.append(dprice.text.replace('₹', ''))
            
        if oprice is None:
            originalprice.append(np.NaN)
        else:
            originalprice.append(oprice.text.replace('₹', ''))
            
        if powerconsump is None:
            powerconsumption.append(np.NaN)
        else:
            powerconsumption.append(powerconsump.text)
            
            
        if rewi is None:
            reviews.append(np.NaN)
        else:
            reviews.append(rewi.text.replace("\xa0"," "))
            
        if score is None:
            rating.append(np.NaN)
        else:
            rating.append(score.text)
        
    
```

```python
#DataFrame
Ac_df = pd.DataFrame({'Product_Name':modelname, 'Original_Price(in Rs)':originalprice,'Discount_Price(in Rs)':discountprice,'Power_consumption(Watts)':powerconsumption,'Capacity':capacity,'Reviews':reviews, 'Rating': rating})
```

```python
Ac_df
```

```python
#Extracting Brand from Product_name.
Ac_df['Brand']=Ac_df["Product_Name"].apply(lambda x:re.findall(r"[A-z]+",x)[0])
```

```python
Ac_df
```

```python
#Extracting Energy_consumption from Product Name.
Ac_df['Energy_consumption']=Ac_df["Product_Name"].apply(lambda x:re.findall("[0-9] Star",x))
Ac_df['Energy_consumption']=Ac_df['Energy_consumption'].apply(lambda x:''.join(x))
```

```python
Ac_df
```

```python
#Extracting Power consumption value.
Ac_df["Power_consumption(Watts)"]=Ac_df["Power_consumption(Watts)"].apply(lambda x:re.findall("[+-]?[0-9]+\.[0-9]+ | [0-9]+",x))
Ac_df['Power_consumption(Watts)']=Ac_df['Power_consumption(Watts)'].apply(lambda x:''.join(x))
```

```python
Ac_df
```

```python
# Capacity Power consumption value.
Ac_df['Capacity']=Ac_df["Capacity"].apply(lambda x:re.findall("[0-9]",x))
Ac_df['Capacity']=Ac_df['Capacity'].apply(lambda x:''.join(x))
```
