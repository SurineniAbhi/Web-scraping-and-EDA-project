# Univariate Analysis

```python
df["Brand"].value_counts().plot.bar()
plt.title('number of models of Air conditioners according to Brands')
plt.ylabel('Count')
plt.xlabel('Brands')
```

The above Graph indicates the number of models of Air conditioners according to Graphs.Hence,we conclude that CARRIER has highest number of models followed by Blue and SAMSUNG.

```python
df["Energy_consumption"].value_counts().plot.pie(autopct="%.2f%%")
```

From the above pie chart,we can conclude that the models with 3 Star Energy Consumption are more in number and 1 Star Energy Consumption are least in number.

```python
sns.countplot(x='Capacity', data=df)
plt.title('Capacity of Air Conditioners')
plt.ylabel('count')
plt.xlabel('Capacity')
plt.xticks(rotation = 90)
plt.show()
```

The above Graph indicates the Capacity of the Air Conditioners and their average count of ecah model.Hence we conclude that the Air Conditioners with 1.5 Ton Capacity have more number of Models.

```python
df["Rating"].value_counts().plot.bar()
plt.title('Ratings of Air conditioners')
plt.ylabel('Count')
plt.xlabel('Rating(out of 5)')
```

The above Graph indicates the Ratings of Air conditioners.Hence we conclude that,the models with a Rating of 4.1 and 4.2 are more in number.

```python
df["Power_consumption(Watts)"].hist(bins = 10)
plt.title('Histogram of Power consumption(Watts) of ACs')
plt.xlabel("Watts")
```

The above graph indicates the histographic analysis of "Power Consumption(Watts)" of air conditioners.Hence we conclude that most of the models range between 750 to 1200 Watts(approx)

```python
sns.boxplot(df["Power_consumption(Watts)"])
plt.title("Boxplot of Power consumption(Watts)")
```

The above figure indicates the Box plot of Power consumption(Watts) of Air conditioners.Hence,we conclude that the average value as observed from the graph is 1000 Watts.The Outliers of the above box plot are above 2000.

```python
sns.distplot(df["Power_consumption(Watts)"],kde=True)
```

The above figure indicates the Probability Density Function of the 'Power consumption(Watts) ' of the air conditioners.Hence we conclude that, All the air conditioners range in between 200 to 3000.The raise in the curve indicates that the models are more in number and started raising from the threshold value 400

```python
df["Original_Price(in Rs)"].hist(bins = 5)
plt.title('Histogram of Original prices of ACs')
plt.xlabel("Price")
```

The above graph indicates the histographic analysis of 'Original Price' of air conditioners.Hence we conclude that most of the models range between 52000 to 68000 rupees

```python
sns.distplot(df["Original_Price(in Rs)"],kde=True)
plt.title('Histogram of Original prices of ACs')
plt.show()
```

The above figure indicates the Probability Density Function of the 'Original prices' of the air conditioners.Hence we conclude that, All the air conditioners range in between 20000 to 100000.The raise in the curve indicates that the models are more in number and started raising from the threshold value 25000(approx).

```python
df["Discount_Price(in Rs)"].hist(bins = 5)
plt.title('Histogram of Discount prices of ACs')
plt.xlabel("Price")
```

The above graph indicates the histographic analysis of 'Discount Price' of air conditioners.Hence we conclude that most of the models range between 32000 to 43000 rupees(approx).

```python
sns.distplot(df["Discount_Price(in Rs)"],kde=True)
plt.title('Histogram of Discount prices of ACs')
plt.show()
```

The above figure indicates the Probability Density Function of the 'Discount prices' of the air conditioners.Hence we conclude that, All the air conditioners range in between 20000 to 80000.The raise in the curve indicates that the models are more in number and started raising from the threshold value 22000(approx).

```python
sns.boxplot(df["Rating"])
plt.title("Boxplot of Rating")
```

The above figure indicates the Box plot of ratings of Air conditioners.Hence,we conclude that the average value as observed from the graph is 4.The Outliers of the above box plot are above 4.5 and below 3.5.

```python
sns.violinplot(df["Rating"])
plt.title("voilin plot of Rating")
```

The above figure indicates the violin plot of ratings of Air conditioners.Hence,we conclude that the average value as observed from the plot is 4.The curves indicates the Outliers which are above 4.5 and below 3.5.

# Bi Variate Analysis

```python
df["Original_Price(in Rs)"].corr(df["Discount_Price(in Rs)"])
```

```python
plt.scatter(df["Original_Price(in Rs)"],df["Discount_Price(in Rs)"])
plt.ylabel("Original_Price(in Rs)")
plt.xlabel("Discount_Price(in Rs)")
```

The above figure indicates the scatter plot with respect to Original price and Discount price.Hence,we conclude that as original price increases,Discount price also increases.

```python
sns.barplot(x='Brand', y="Discount_Price(in Rs)", data=df,ci=None)
plt.xticks(rotation=90)
plt.show()
```

The above figure indicates the bar plot of the comparision between Brands with respect to discount prices.Hence we conclude that,the brand named 'O' have highest discounted price of >60000(approx)

```python
sns.barplot(x='Brand', y='Rating', data=df,ci=None)
plt.xticks(rotation = 90)
plt.show()
```

The above figure indicates the bar plot for Brand and Average Rating.Hence we conclude that,the brand named 'ONIDA' have highest average Rating >4

```python
#extracting the count values of Brands with respect to energy consumption
crosstab = pd.crosstab(df['Brand'], df['Energy_consumption'])
crosstab
```

```python
crosstab.plot(kind="bar", 
                 stacked=True)
```

The above figure indicates the Crosstab plot of Brands and energy Consumptions of air conditioners.Hence we conclude that,SAMSUNG has 3 Star energy consumption of 32 models,4 Star energy consumption of about 5 models and 18 Star energy consumption of about 17 models(approx)

```python
sns.lineplot(x = "Rating", y = "Energy_consumption", data = df)
```

The above figure indicates the line plot with respect to energy consumption and rating.Hence we can conclude that most of the air conditioners with rating of 2.5 to 3.5 have 5 Star Energy consumption.

```python
plt.scatter(df["Power_consumption(Watts)"],df["Energy_consumption"])
plt.ylabel('Energy_consumption')
plt.xlabel('Power_consumption(Watts)')
```

The above figure indicates the scatter plot with respect to Power Consumption and Energy Consumption.
